```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
metricas = ['flow_rate','pressure_point','direct_play','counterattack','T_zona_media','shots','T_build_up','crossing']


colsnames = {"direct_play":"Direct play", 
        "counterattack": "Counterattack",
        "T_build_up": "Build up",
        "T_zona_media": "Midfield play",
        "flow_rate":"Flow rate",
        "crossing":"Crossing",
        "pressure_point":"Pressure point",
        "shots":"Shots",

}
```

```python
arx = './metricas_agregadas_con_log.csv'
df_log = pd.read_csv(arx)
```

```python
arx = './metricas_agregadas_sin_log.csv'
df = pd.read_csv(arx)
```

```python
import statsmodels.api as sm


for met in metricas:

    x1 = df[met]
    x2 = df_log[met]

    # 1. Calcular asimetría original (con cola a la derecha será > 0)
    asimetria_original = stats.skew(x1)

    # 2. Calcular asimetría después del logaritmo
    asimetria_transformada = stats.skew(x2)

    print(f"Asimetría original: {asimetria_original:.4f}")
    print(f"Asimetría post-logaritmo: {asimetria_transformada:.4f}")


    fig,ax = plt.subplots(1,2)

    ax[0].hist(x1)
    ax[1].hist(x2)

    plt.show()


```
