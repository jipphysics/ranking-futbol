```python
import json,os
import numpy as np
import pandas as pd
```

### Cargo datos

```python
ligue2team2rank = json.load( open("../league2team2rank.json", "r") )
```

```python
matches = pd.read_csv("../matches.csv")
matches.head(2)
```

```python
metrics = pd.read_csv("metricas_hasta_Fecha3.csv")
metrics.head(2)
```

### Armo redes

Primero cruzo los datos por match

```python
df0 = pd.merge(metrics, matches[['match', 'home', 'away', 'winner', 'score_home','score_away']], on='match', how='left')

df0.head(2)
```

```python
metricas = ['direct_play','counterattack','T_maintenance','T_build_up',
            'T_zona_media','flow_rate','crossing','pressure_point','pressure_loss',	
            'shots']

ligas = ['England','France','Germany','Italy','Spain']


for liga in ligas:

    print(liga)

    carpeta = f'redes_hasta_fecha3/{liga}'
    if not os.path.exists(carpeta):
        os.makedirs(carpeta)


    team2rank = ligue2team2rank[liga]
    df = df0[ df0['liga']==liga ]


    for metrica in metricas:

        print(f"\t{metrica}")

        d={}
        d2={}

        def f(row):

            # tomo datos
            team1 = str(row.team1)
            nodo1 = team2rank[team1]

            if row.team1==row.home:
                team2=str(row.away)
            else:
                team2=str(row.home)

            nodo2 = team2rank[team2]

            w = row[metrica] 

            try:
                d[(nodo1,nodo2)]+=w
            except:
                d[(nodo1,nodo2)]=w


            try:
                d2[(team1,team2)]+=w
            except:
                d2[(team1,team2)]=w
            

        df.agg( lambda x: f(x), axis=1 )


        # armo red
        Nmax = max(elemento for tupla in d for elemento in tupla)
        red = []

        for i in range(1,Nmax):

            for j in range(i+1, Nmax+1):
                
                try:
                    fij = d[(j,i)] - d[(i,j)] 

                    red.append( [i,j, fij] )
                except:
                    None



        #guardo data
        arx_out = f'{carpeta}/red_{metrica}.dat'

        np.savetxt(
            arx_out,               
            red,               
            fmt='%d,%d,%.4e',         
            delimiter=',', 
            header='n1,n2,fij',
            comments=''
        )
```
